```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
#function to extract product title
def get_title(soup):
    try:
        title=new_soup.find("h1", attrs={"class":'pdp-e-i-head'}).text.strip()
        
    except AttributeError:
        title=''
        
    return title
```

```python
def get_price(soup):
    try:
        price=new_soup.find("span", attrs={"class":'pdp-final-price'}).find("span", attrs={"class":'payBlkBig'}).text
        
    except AttributeError:
        price=''
        
    return price
```

```python
def get_rating(soup):
    try:
        rating=new_soup.find("span", attrs={"itemprop":'ratingValue'}).text.strip()
        
    except AttributError:
        rating=''
        
    return rating
```

```python
def rating_count(soup):
    try:
        rating_count=new_soup.find("span", attrs={"itemprop":'ratingCount'}).text.strip()
        
    except AtrributeError:
        rating_count=''
        
    return rating_count
```

```python
if __name__ == '__main__':
    
    #add user agent
    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})
    
    #webpage url
    URL = "https://www.snapdeal.com/products/mens-tshirts-polos?sort=plrty"
    
    #http request
    webpage = requests.get(URL, headers=HEADERS)
    
    # Soup Object containiang all data
    soup = BeautifulSoup(webpage.content, "html.parser")
    
    #Fetch links as list of tags objects
    links = soup.find_all("a", attrs={'class':'dp-widget-link noUdLine'})
    
    #store the links
    links_list=[]
    
    #loop for extracting links from tag objects
    for link in links:
        links_list.append(link.get('href'))
        
    d={"title":[],"price":[],"rating":[],"rating_count":[]}
    
    # Loop for extracting product details from each link 
    for link in links_list:
        new_webpage = requests.get(link, headers=HEADERS)
        
        new_soup = BeautifulSoup(new_webpage.content, "html.parser")
        
        # Function calls to display all necessary product information
        d['title'].append(get_title(new_soup))
        d['price'].append(get_price(new_soup))
        d['rating'].append(get_rating(new_soup))
        d['rating_count'].append(rating_count(new_soup))
        
        snap_df = pd.DataFrame.from_dict(d)
        snap_df['title'].replace('', np.nan, inplace=True)
        snap_df = snap_df.dropna(subset=['title'])
        snap_df.to_csv("snapdeal_data.csv", header=True, index=False)
        
```
