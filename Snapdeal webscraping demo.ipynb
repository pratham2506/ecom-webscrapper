```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
URL = "https://www.snapdeal.com/products/mens-tshirts-polos?sort=plrty"
```

```python
# Headers for request
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'}) #add your user agent 
```

```python
# HTTP Request
webpage = requests.get(URL, headers=HEADERS)
```

```python
type(webpage.content)
```

```python
# Soup Object containiang all data
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
soup
```

```python
#Fetch links as list of tags objects
links = soup.find_all("a", attrs={'class':'dp-widget-link noUdLine'})
```

```python
link=links[0].get('href')
```

```python
link
```

```python
new_webpage = requests.get(link, headers=HEADERS)
```

```python
new_webpage
```

```python
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
```

```python
new_soup
```

```python
new_soup.find("h1", attrs={"class":'pdp-e-i-head'}).text.strip()
```

```python
new_soup.find("span", attrs={"class":'pdp-final-price'}).find("span", attrs={"class":'payBlkBig'}).text
```
